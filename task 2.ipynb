```python
# Titanic EDA with Seaborn
# Matplotlib is a low-level library for creating basic visualizations,
# while Seaborn is a high-level library built on Matplotlib that makes statistical plots more attractive and easier to create.
# 1. Seaborn Library

# Definition: Seaborn is a Python data visualization library built on top of Matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.

# In Task: You used Seaborn to create plots (countplot, histplot, heatmap) that reveal survival patterns in the Titanic dataset.

# 2. Handling Missing Values

# Definition: Missing value imputation replaces null values to avoid errors in visualization and analysis.

# In Task: Median was used to fill missing Age values (numeric), and mode was used for Embarked (categorical).

# 3. Count Plot (Survival Distribution)

# Definition: A count plot shows the frequency of each category in a dataset using bars.

# In Task: The count plot displayed how many passengers survived (1) vs did not survive (0).

# 4. Histogram (Age vs Survival)

# Definition: A histogram displays the distribution of numerical data divided into intervals (bins). Seaborn allows grouping with hue to compare categories.

# In Task: The histogram showed how survival chances varied across different age groups (e.g., children vs adults).

# 5. Heatmap (Correlation Matrix)

# Definition: A heatmap is a color-coded representation of a 2D matrix, often used to visualize correlations between variables.

# In Task: The heatmap displayed correlations among numerical features (like Age, Fare, and Survival), helping to identify which features influence survival.

#  Summary:
# This task used Seaborn visualizations to:

# Clean and prepare the Titanic dataset.

# Show overall survival distribution.

# Compare age distribution between survivors and non-survivors.

# Highlight relationships between variables with a heatmap.
```

```python
import pandas as pd, seaborn as sns, matplotlib.pyplot as plt
url = "https://raw.githubusercontent.com/dsrscientist/dataset1/master/titanic_train.csv"
df = pd.read_csv(url)
# Fill missing values
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
# Quick EDA plots
sns.countplot(x='Survived' , data=df)
plt.title("Survival Distribution"); plt.show()
sns.histplot(data=df, x='Age', hue='Survived', multiple="stack")
plt.title("Age vs Survival"); plt.show()
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("correlation heatmap"); plt.show()

                      
```
