```python
# Explanation with Definitions
# 1. Seaborn Dataset

# Definition: Seaborn is a visualization library in Python that also provides some built-in datasets (like Titanic, Tips, Iris).

# In our task, we used the Titanic dataset.

# It contains passenger details like sex, age, fare, class and whether they survived or not.

# 2. Data Cleaning

# Real-world datasets contain missing values.

# We removed rows with missing age, fare, sex, or class.

# We also converted categorical values (like male/female, class) into numeric format using dummy variables (one-hot encoding).

# 3. Feature (X) and Target (y)

# Feature (X): The input variables used to predict (example: age, fare, sex, class).

# Target (y): The variable we want to predict (survived: 0 = not survived, 1 = survived).

# 4. Train-Test Split

# Definition: Splitting data into training and testing sets.

# Training set: Used to train the model (learn patterns).

# Testing set: Used to check model accuracy.

# Here, we used 80% data for training and 20% for testing.

# 5. Decision Tree Classifier

# Definition: A Decision Tree is a machine learning algorithm that splits data into branches based on conditions until it reaches a decision.

# Example:

# If sex = female → higher chance of survival.

# If fare > 30 → higher chance of survival.

# It works like a flowchart of if-else conditions.

# 6. Model Training

# The model learns from the training data by finding the best splits (based on features).

# In our case, it looked at age, fare, sex, and class to decide survival.

# 7. Accuracy

# Definition: Accuracy = Correct Predictions / Total Predictions.

# Tells us how well the model performed on unseen test data.

# 8. Decision Tree Visualization

# A tree diagram is plotted showing how the model makes decisions step by step.

# Each node shows a condition (like fare <= 23.35).

# Each leaf node shows the prediction (Survived or Not).
```

```python
# IMPORT LIBRARIES
import seaborn as sns
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
# Load Titanic dataset from seaborn
df = sns.load_dataset("titanic")
# Drop rows with missing values in key columns
df = df[["sex", "age", "fare", "class", "survived"]].dropna()
# Convert categorical columns into numeric
df = pd.get_dummies(df, drop_first=True)
# Features (X) and Target (y)
x = df.drop("survived", axis=1)
y = df["survived"]
# Train-test split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
# Decision Tree Classifier
clf = DecisionTreeClassifier(max_depth=3, random_state=42)
clf.fit(x_train, y_train)
# Accuracy
print("Accurcy:",clf.score(x_test, y_test))
# Plot Decision Tree
plt.figure(figsize=(12,6))
plot_tree(clf, feature_names=x.columns, class_names=["Not Survived", "survived"], filled=True)
plt.show()

```
